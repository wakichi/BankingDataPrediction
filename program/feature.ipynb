## age

顧客の年齢のデータ. (int, 18~95)
- EDAして良さそうな情報見つけたい。
- ２０代、30代などのカテゴリ変数を作る。＝＞分け方を工夫
- 他のカテゴリ変数との関係を探す(interaction)

```python
def campaign_binify(x):
    if 18<= x <=22:
        return 1
    if 23<= x <= 30:
        return 2
    if 31<= x <= 61:
        return 3
    if 61< x:
        return 4
```

```python
df["age_bin"] = df.age.map(campaign_binify)
```

## duration

最終のときに接触していた時間 
- （仮説）大きいほどyesの割合高いのでは？
=> yesとnoでグラフの形、もしくはmeanを見てみる
=> mean yes:539, no: 221　stdもまあまあ小さいので信頼できそう。
- 左に寄った形をしているのでlog取りたいお気持ち
- 

```python
df["duration"] = df.duration.map(lambda x: x+1)
df["duration_log"] = np.log(df.duration)
df["duration_sqrt"] = np.sqrt(df.duration)
```

```python
save_feature(df)
```

## balance

貯金残高
- 多いほどyesの平均が高い。　上振れのものは圧倒的にyが高い。
＝＞グラフで比較してみるのもあり
- ハズレ値を除去する
＝＞testにある場合はどうする？いくつある？閾値は？
=> 8.6以下、10.3以上は外れ値と認定して0にする。  
testデータには18個しないので問題なし。
- logにする

```python
min_balance = abs(min(df.balance))
min_balance
```

```python
df["balance"] = df.balance.map(lambda x: x+min_balance+1)
df["balance_log"] = np.log(df.balance)
df["balance_sqrt"] = np.sqrt(df.balance)
```

```python
save_feature(df)
```
